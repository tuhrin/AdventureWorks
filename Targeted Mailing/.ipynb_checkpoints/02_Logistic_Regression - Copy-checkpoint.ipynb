# Logistic Regression

```python
# General 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Logistic regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# ROC curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

%matplotlib inline
```

```python
# Import data to DataFrame
data=pd.read_csv('DATA/TargetedMailing.csv')
data.head()
```

```python
# Categorical and Numerical variables
data.Buyer=data.Buyer.astype(object) # set Buyer as object

cat_cols=data.select_dtypes(include=['object']).columns.tolist()
num_cols=data.select_dtypes(exclude=['object']).columns.tolist()

print('Categorical variables:',cat_cols)
print('Numerical variables:',num_cols)
```

```python
# create dummy variables
for cat in cat_cols:
    cat_list = 'var_'+cat
    cat_list = pd.get_dummies(data[cat], prefix=cat, drop_first='True')
    data_dummy=data.join(cat_list)
    data=data_dummy

keep_cols=[x for x in data.columns.tolist() if x not in cat_cols]
data=data[keep_cols]
data.head()
```

```python
# Define input and output variables
y=data.columns[-1]
x=data.columns[:-1]
```

### Logistic regression with Feature selection - RFE (Recursive Feature Elimination)

```python
# Feature selection
from sklearn.feature_selection import RFE

logreg = LogisticRegression()
rfe = RFE(logreg)
rfe = rfe.fit(data[x], data[y])

print(rfe.support_)
print(rfe.ranking_)
```

```python
# get list of columns to keep
x_rfe=x[rfe.support_]
print('Columns to keep: ',x_rfe.tolist())
```

```python
#correlation matrix
corrmat = data[x_rfe].corr()
f, ax = plt.subplots(figsize=(9, 9))
sns.heatmap(corrmat, square=True, annot=True, fmt='.2f');
```

```python
# train model
X_train, X_test, y_train, y_test = train_test_split(data[x_rfe], data[y], test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# accuracy
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.3f}'.format(logreg.score(X_test, y_test)))
```

```python
# confussion matrix
metrics.confusion_matrix(y_test, y_pred)
```

```python
# for ROC curve
logit_roc_auc_rfe = roc_auc_score(y_test, logreg.predict(X_test))
fpr_rfe, tpr_rfe, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
```

### Logistic regression with Feature Selection - Univariate Selection / SelectKBest

```python
# Feature Selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
test = SelectKBest(score_func=chi2, k=6)
fit = test.fit(data[x], data[y])
np.set_printoptions(precision=5)
print(fit.scores_)
```

```python
# get list of columns to keep
x_us=x[np.argpartition(fit.scores_, 6)[-6:]]
print('Columns to keep: ',x_us.tolist())
```

```python
#correlation matrix
corrmat = data[x_us].corr()
f, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(corrmat, square=True, annot=True, fmt='.2f');
```

```python
# train model
X_train, X_test, y_train, y_test = train_test_split(data[x_us], data[y], test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# accuracy
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.3f}'.format(logreg.score(X_test, y_test)))
```

```python
# confussion matrix
metrics.confusion_matrix(y_test, y_pred)
```

```python
# for ROC curve
logit_roc_auc_us = roc_auc_score(y_test, logreg.predict(X_test))
fpr_us, tpr_us, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
```

### Logistic regression with none Feature Selection

```python
#correlation matrix
corrmat = data[x].corr()
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, square=True);
```

```python
# train model
X_train, X_test, y_train, y_test = train_test_split(data[x], data[y], test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# accuracy
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.3f}'.format(logreg.score(X_test, y_test)))
```

```python
# confussion matrix
metrics.confusion_matrix(y_test, y_pred)
```

```python
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
```

#### Receiver operating characteristic

```python
plt.figure(figsize=(12,9))
plt.plot(fpr_rfe, tpr_rfe, label='Logistic Regression with Recursive Feature Elimination (area = %0.3f)' % logit_roc_auc_rfe)
plt.plot(fpr_us, tpr_us, label='Logistic Regression with Univariate Selection (area = %0.3f)' % logit_roc_auc_us)
plt.plot(fpr, tpr, label='Logistic Regression with none Feature Selection (area = %0.3f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
